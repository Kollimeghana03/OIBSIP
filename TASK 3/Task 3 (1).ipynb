# Visualizing The Dataset

```python
sns.set_style("whitegrid")
plt.figure(figsize=(7, 5))
sns.histplot(data.price, kde=True)  
plt.show()
```

```python
x = data.drop(['price', 'CarName', 'fueltype', 'aspiration',	'doornumber', 'carbody',	'drivewheel', 'enginelocation', 'enginetype', 'cylindernumber', 'fuelsystem'], axis=1)
y = data['price']
```

# Car Price Prediction Model

```python
# Importing train test split
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=1234)
```

```python
# Importing Linear Regression library
from sklearn.linear_model import LinearRegression

model = LinearRegression()
model.fit(x_train,y_train)
```

```python
y_pred = model.predict(x_train)
y_pred
```

```python
from sklearn import metrics

error_square = metrics.r2_score(y_train, y_pred)
print("Error square: ", error_square)
```

```python
plt.scatter(y_train,y_pred)
plt.xlabel('Actual price')
plt.ylabel('Predicted price')
plt.title("Actual vs Predicted")
plt.show()
```
